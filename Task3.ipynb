### Task 3: Data visualization

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df= pd.read_csv('books_data.csv')
```

```python
df['Price']=df['Price'].str.replace("Â£","").astype(float)
```

```python
plt.figure(figsize=(7,5))
sns.histplot(df["Price"], bins=8)
plt.title("Book Price Distribution")
plt.xlabel("Price")
plt.show()
```

```python
top = df.sort_values("Price", ascending=False).head(10)

plt.figure(figsize=(8,5))
sns.barplot(x=top["Price"], y=top["Title"])
plt.title("Top 10 Expensive Books")
plt.show()
```

```python
plt.figure(figsize=(5,4))
sns.countplot(x="Availability", data=df)
plt.title("Availability Status")
plt.show()
```
